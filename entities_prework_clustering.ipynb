```python
law_keep = [
    "the Paris Agreement",
    "a Memorandum of Understanding",
    "the Clean Air Act",
    "Vision 2030",
    "the National Environmental Policy Act",
    "the Petroleum Industry Act",
    "Paris Agreement"
]

product_keep = [
    "Cybertruck", "Model Y", "Model 3", "Model S",
    "Optimus", "Autopilot", "EVs", "Full Self-Driving"
]

law_filtered = filtered_df[
    (filtered_df['label'] == 'LAW') &
    (filtered_df['entity'].str.contains('|'.join(law_keep), regex=True))
]

product_filtered = filtered_df[
    (filtered_df['label'] == 'PRODUCT') &
    (filtered_df['entity'].isin(product_keep))
]

other_filtered = filtered_df[
    ~filtered_df['label'].isin(['LAW', 'QUANTITY', 'PRODUCT'])
]

final_filtered_df = pd.concat([law_filtered, product_filtered, other_filtered], ignore_index=True)

final_filtered_df.head()
```

```python
final_filtered_df.sort_values(by="count", ascending=False).reset_index(drop=True)
```

```python
# group by entity and sum their counts
entity_counts = final_filtered_df.groupby('entity')['count'].sum().reset_index()

# total mentions overall
total_mentions = entity_counts['count'].sum()

# compute percentage contribution per entity
entity_counts['percent'] = (entity_counts['count'] / total_mentions) * 100

# sort by total mentions descending
entity_counts = entity_counts.sort_values('count', ascending=False).reset_index(drop=True)

# cumulative percentage to find coverage thresholds (like 90%, 95%)
entity_counts['cumulative_percent'] = entity_counts['percent'].cumsum()

entity_counts.head(20)
```

```python
import torch
from sentence_transformers import SentenceTransformer, util
```

```python
# Entity types to cluster
types_to_cluster = ['PERSON', 'GPE', 'ORG', 'FAC', 'LOC', 'EVENT', 'WORK_OF_ART']

# Load model once
device = "mps" if torch.backends.mps.is_available() else "cpu"
print(f"Using device: {device}")
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2", device=device)

# Dictionary to store all mappings
entity_mapping = {}

# Process each entity type
for entity_type in types_to_cluster:
    print(f"\n{'='*50}")
    print(f"Processing {entity_type}...")
    print(f"{'='*50}")

    # Get unique entities for this type
    entities_list = df_to_cluster[df_to_cluster['label'] == entity_type]['entity'].unique().tolist()
    print(f"Unique entities: {len(entities_list)}")

    if len(entities_list) < 2:
        print(f"Skipping {entity_type} (too few entities)")
        continue

    # Create embeddings
    print("Creating embeddings...")
    embeddings = model.encode(
        entities_list,
        convert_to_tensor=True,
        normalize_embeddings=True,
        device=device,
        show_progress_bar=True
    )

    # Run community detection
    print("Running clustering...")
    threshold = 0.78
    min_community_size = 2

    clusters = util.community_detection(
        embeddings,
        threshold=threshold,
        min_community_size=min_community_size
    )

    print(f"Found {len(clusters)} clusters")

    # Create mapping from clusters
    for cluster in clusters:
        # Find representative, the most central entity
        cluster_indices = torch.tensor(cluster, dtype=torch.long, device=embeddings.device)
        cluster_embeddings = embeddings[cluster_indices]
        centroid = cluster_embeddings.mean(dim=0, keepdim=True)
        similarities = util.cos_sim(centroid, cluster_embeddings).squeeze(0)
        rep_idx = cluster[int(torch.argmax(similarities).item())]
        representative = entities_list[rep_idx]

        # Map all cluster members to representative
        for idx in cluster:
            entity_mapping[entities_list[idx]] = representative

    # sample clusters
    print(f"\nSample clusters for {entity_type}:")
    for i, cluster in enumerate(clusters[:3]):  # first 3 clusters
        members = [entities_list[idx] for idx in cluster]
        print(f"  Cluster {i+1}: {members}")

print(f" Total mappings created: {len(entity_mapping)}")
```

```python
# Apply the mappings to the dataframe
print("Applying entity mappings to dataframe...")

# Replace entities using the mapping
df_to_cluster['entity_original'] = df_to_cluster['entity']  # keep original for reference
df_to_cluster['entity'] = df_to_cluster['entity'].map(entity_mapping).fillna(df_to_cluster['entity'])

# Check how many were replaced
replaced_count = (df_to_cluster['entity'] != df_to_cluster['entity_original']).sum()
print(f" Replaced {replaced_count:,} entity names")

# checking examples
print("\nExample replacements:")
examples = df_to_cluster[df_to_cluster['entity'] != df_to_cluster['entity_original']][['entity_original', 'entity', 'label']].head(10)
print(examples)
```

```python
# Check PERSON replacements
print("PERSON entity replacements:")
person_examples = df_to_cluster[
    (df_to_cluster['entity'] != df_to_cluster['entity_original']) &
    (df_to_cluster['label'] == 'PERSON')
][['entity_original', 'entity', 'label']].drop_duplicates().head(20)
print(person_examples)

print("\n" + "="*50)

# Check ORG replacements
print("\nORG entity replacements:")
org_examples = df_to_cluster[
    (df_to_cluster['entity'] != df_to_cluster['entity_original']) &
    (df_to_cluster['label'] == 'ORG')
][['entity_original', 'entity', 'label']].drop_duplicates().head(20)
print(org_examples)
```
